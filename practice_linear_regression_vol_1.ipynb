```python
import pandas as pd
import numpy as np
import sklearn.model_selection as ms
import matplotlib.pyplot as plt
# Baca data
df1 = pd.read_csv('bensin.csv')
liter=df1[['Liter']]
kilometer = df1[['Kilometer']]
# Split data
X_train, X_test, y_train, y_test = ms.train_test_split(liter, kilometer, test_size=0.2, random_state=0)

# Garis regresi manual (berdasarkan model sebelumnya)
xl = np.linspace(0, 45)
yl = 10.641 + 6.458 * xl

#buat grafik
plt.figure(figsize=(8, 6))
plt.scatter(X_train, y_train, edgecolors='r', label='Data Asli')
plt.plot(xl, yl, color='blue', label='Garis Regresi')

#tambah visual
plt.xlabel('Liter')
plt.ylabel('Kilometer')
plt.title('Konsumsi Bahan Bakar')
plt.legend()
plt.grid(True)

plt.show()
```

```python
r2 = model1.score(X_test, y_test)
print(r2)
```

```python
import sklearn.linear_model as lm
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Buat dan latih model
model1 = lm.LinearRegression()
model1.fit(X_train, y_train)

# Hitung skor R^2
r2 = model1.score(X_test, y_test)
print("RÂ² Score:", round(r2, 2))

# Prediksi di data test
y_pred = model1.predict(X_test)

# Hitung MAE dan MSE
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print("Mean Absolute Error (MAE):", round(mae, 2))
print("Mean Squared Error (MSE):", round(mse, 2))
```
